```python
# グラフを書く準備
import numpy as np
import matplotlib.pyplot as plt

# 0から1までを100刻み
# t = np.linspace(0, 1, 100)

x = np.linspace(-5, 5, 100)


# 図を描く宣言
plt.figure(0)

# 実際のプロット（図に描く本文，本体．横軸：ｔ，縦軸：8*t-4*t**2）
# plt.plot(t, 8*t-4*t**2, color="red")
# plt.plot(t, np.sin(t), color="black")
# plt.plot(t, np.exp(t), color="green")
plt.plot(x, x**2+3)
plt.plot(x, 2*x+10)

# タイトル
plt.title('graph', {'fontsize':100}) 

# X軸ラベル
plt.xlabel('x', {'fontsize':15}) 

# Y軸ラベル
plt.ylabel('f(x)', {'fontsize':15}) 

# 画面に表示
plt.show()
```
